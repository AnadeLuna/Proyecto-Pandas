# Analise the Global Shark Attaks.

### 1º: Import all the libraries that i will need use to analise.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

### 2º: Import the DataSet from datas_clean.csv.

```python
df = pd.read_csv("../data/datas_clean.csv")
```

### 3º: Check the data.

```python
#Number of cases:
df.shape
```

```python
df.columns
```

### Shark Attacks in the last 10 years.

```python
plt.figure(figsize=(8,6))
sns.countplot(x="Year",data=df, palette="husl")
```

### Shark Attacks in the last 10 years per months.

```python
#Here I would like to order in months.But I couln´t.
plt.figure(figsize=(15,10))
sns.countplot(x="Month",hue="Year",data=df, palette="Set2")
```

### Shark Attacks in the last 10 years per countries.

```python
plt.figure(figsize=(22,17))
sns.scatterplot(x="Year",y="Country", data=df, palette="Greens")
```

### Shark Attacks in the last 10 years per sex.

```python
plt.figure(figsize=(8,6))
sns.countplot(x="Year",hue="Sex", data=df, palette="Blues")
```

### Shark Attacks in the last 10 years per Fatal.

```python
plt.figure(figsize=(8,6))
sns.countplot(x="Year",hue="Fatal", data=df, palette="Reds")
```
